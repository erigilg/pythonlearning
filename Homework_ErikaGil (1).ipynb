```
import pandas as pd
from vega_datasets import data
carsdf = data.cars()
```

```
carsdf.head()
```

```
carsdf.plot.scatter(x='Miles_per_Gallon',y='Cylinders', title=' MPG by Cylinders')
```

```
import matplotlib.pyplot as plt
```

```
plt.scatter(carsdf.Miles_per_Gallon,carsdf.Cylinders, s=30,alpha=0.3, c='r')
plt.xlabel('MPG')
plt.ylabel('Cylinders')
plt.title('Scatter plot on MPG vs Cylinders')
plt.show()
```

```
fig = plt.figure()
graph = fig.add_subplot(111)
graph.scatter(carsdf.Weight_in_lbs,carsdf.Acceleration, s=40,alpha=0.3, c='b')
graph.xlabel('Weight (lbs)')
graph.ylabel('Acceleration')
graph.title('Scatter plot on Weight vs Acceleration')
graph.show()
```
